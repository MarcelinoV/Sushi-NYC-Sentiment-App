```python
# convert above cell into function

def base_transformer(text):

    encoded_example = base_tokenizer(text, return_tensors='pt')
    output = base_model(**encoded_example)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)

    scores_dict = {
        'base_neg':scores[0],
        'base_neu':scores[1],
        'base_pos':scores[2]
    }

    return scores_dict

def foody_transformer(text):

    encoded_example = foody_tokenizer(text, return_tensors='pt')
    output = foody_model(**encoded_example)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)

    scores_dict = {
        'foody_neg':scores[0],
        'foody_neu':scores[1],
        'foody_pos':scores[2]
    }

    return scores_dict

def absa_transformer(text):
    output = absa_model(text)
    # scores = output[0][0].detach().numpy()
    # scores = softmax(scores)

    # scores_dict = {
    #     'absa_neg':scores[0],
    #     'absa_neu':scores[1],
    #     'absa_pos':scores[2]
    # }

    return output
```

```python
# build dataset

res = {}
base_keys = ['base_neg', 'base_neu', 'base_pos']
food_keys = ['foody_neg', 'foody_neu', 'foody_pos']

for row in sushi_reviews.iterrows():
    
    text = row[1]['TEXT']
    id = row[1]['REVIEW_ID']
    vader_result = sia.polarity_scores(text)
    vader_result_rename = {}
    for key, val in vader_result.items():
        vader_result_rename[f'vader_{key}'] = val

    try:

        roberta_base_result = base_transformer(text)
        roberta_foody_result = foody_transformer(text)
        sentiment_results = vader_result_rename | roberta_base_result | roberta_foody_result
        res[id] = sentiment_results
    except RuntimeError:
        roberta_base_result = dict.fromkeys(base_keys)
        roberta_foody_result = dict.fromkeys(food_keys)
        sentiment_results = vader_result_rename | roberta_base_result | roberta_foody_result
        res[id] = sentiment_results
        
        print('Adjusted for ID {}'.format(id))
    # break
```

```python
# create word tokens

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

def gen_word_tokens(df, text):
    # Load string
    # raw_data = pd.read_csv(file + '.csv')
    
    df['dummy'] = df[text].astype(str)
    # Convert to lowercase
    dummy_gen = (post.lower() for post in df['dummy'])
    df['dummy'] = [i for i in dummy_gen]

    # Word & Sentence Tokenization
    token_post = (word_tokenize(post) for post in df['dummy'])
    token_post = [i for i in token_post]

#     token_post = [sent_tokenize(post) for post in token_post]
    
    #df['token_post'] = token_post
    
    # Remove Punctuation
    reg = re.compile(r'(@[a-z0-9]+)|([^0-9a-z \t])|(\w+:\/\/\S+)')

    no_punc = []

    for filt in token_post:
        review = []
        for token in filt:
            new_token = reg.sub(u'', token)
            if not new_token == u'':
                review.append(new_token)
        no_punc.append(review)
        
#     df['token_post'] = no_punc
        
    # Remove Stopwords
    no_stop = []

    for post in no_punc:
        new_term_vector = []
        for word in post:
            if not word in stopwords.words('english'):
                new_term_vector.append(word)

        no_stop.append(new_term_vector)
        
#     df['token_post'] = no_stop
        
    # Stemming & Lemmatization
    pstem = PorterStemmer()
    wlem = WordNetLemmatizer()

    preproc_text = []

    for text in no_stop:
        final_text = []
        for word in text:
            pstem.stem(word)
            final_text.append(wlem.lemmatize(word))

        preproc_text.append(final_text)
        
    #df['token_post'] = preproc_text
        
    # create final data set
    #data = df.copy()

    #new_col = pd.Series(preproc_text)
    df['WORD_TOKENS'] = [list(set(i)) for i in preproc_text]
    df.drop('dummy', axis=1, inplace=True)
    return df['WORD_TOKENS'].tolist()
```
