```python
# Further processing

import polars as pl
import json

# Load the JSON data
# with open("/mnt/data/google_place_example.txt", "r") as file:
#     data = json.load(file)

places = data["places"]  # Assuming the JSON has a "places" key

# Keys to exclude
exclude_keys = {
    "reviews",  # Reviews will go into the second dataset
    "googleMapsUri", "websiteUri",  # Long URLs/URIs
    "regularOpeningHours", "currentOpeningHours", "currentSecondaryOpeningHours", "regularSecondaryOpeningHours", 
    "photos", "types",  # Hours-related keys
    "addressComponents", "adrFormatAddress", "generativeSummary", "addressDescriptor", "googleMapsLinks", "containingPlaces"
}

# Function to flatten specific nested keys
def flatten_nested_data(place):
    flattened = {}
    for key, value in place.items():
        if key == "priceRange" and isinstance(value, dict):
            # Extract startPrice and endPrice values if they exist
            start_price = value.get("startPrice", {}).get("units")
            end_price = value.get("endPrice", {}).get("units")
            flattened["startPrice"] = start_price
            flattened["endPrice"] = end_price
        elif key in {"addressComponents", "location", "viewport", "displayName", "primaryTypeDisplayName", 
                     "editorialSummary", "paymentOptions", "parkingOptions", "accessibilityOptions"} and isinstance(value, dict):
            # Expand other nested dictionaries into new columns
            for sub_key, sub_value in value.items():
                flattened[f"{key}_{sub_key}"] = sub_value
        elif key not in exclude_keys:  # Exclude specific keys
            flattened[key] = value
    return flattened

# Create the metadata DataFrame
metadata_records = [flatten_nested_data(place) for place in places]
metadata_df = pl.DataFrame(metadata_records)

lang_code_columns = [i for i in metadata_df.columns if 'languageCode' in i]

metadata_df = metadata_df.drop(lang_code_columns)

# Create the reviews DataFrame
reviews_data = []
for place in places:
    place_id = place["id"]
    reviews = place.get("reviews", [])
    for review in reviews:
        review["place_id"] = place_id  # Add place_id as a foreign key
        reviews_data.append(review)

reviews_df = pl.DataFrame(reviews_data)

# Save the DataFrames
# metadata_df.write_parquet("metadata.parquet")
# reviews_df.write_csv("reviews.csv")
```
